# cancel an invoice

Canceling an invoice keeps it in the system?

* in squareup dashboard the invoice is shown as canceled
* in the squareup api call above it will show as canceled

```python
invoice_id = "inv:0-ChCihotCgtMe0_3c736v9N7MEMoL"

result = sq_client.invoices.cancel_invoice(
  invoice_id = invoice_id,
  body = {
    "version": 0
  }
)

if result.is_success():
  print(result.body)
elif result.is_error():
  print(result.errors)
```

```python
pp.pprint(result.body)
```

When you list the orders, the field 'id' which comes in the search order api output right after fullfillments is the oder id. In this case, 'rSyA0suDbb9wn0DDe6wWgQd2DLYZY'.

However, when you attempt to create an order invoice, it fails because we created this via the store and the shopping cart has already generated an invoice?
