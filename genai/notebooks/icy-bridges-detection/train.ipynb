```python
from ultralytics import YOLO

# Training of a model
model = YOLO("yolov8n.yaml")

result = model.train(
    data="config.yaml",
    epochs=2000,
    patience=0,
    optimizer= "SGD",
    cos_lr = True
)

```

```python
# Validate the trained model
metrics = model.val(max_det=3)
```
